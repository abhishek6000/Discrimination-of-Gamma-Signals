# Import

```python
from sklearn.metrics import accuracy_score
from sklearn import tree
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import svm
from sklearn.utils import shuffle
```

```python
data = pd.read_csv('/kaggle/input/magic-gamma-telescope-dataset/telescope_data.csv')
```

```python
data = shuffle(data)
```

# Train-Test Load

```python
train_mat = data.iloc[0:15000,:] # 0 theke 15000 toh.?
train_labels = train_mat['class']
train_mat = train_mat.drop(['Unnamed: 0','class'], axis=1)
train_mat = train_mat.values

test_mat = data.iloc[15000:,:]
test_labels = test_mat['class']
test_mat = test_mat.drop(['Unnamed: 0','class'], axis=1)
test_mat = test_mat.values
```

```python
test_mat.shape
```

```python
train_Y = []
test_Y = []
for l in train_labels:
    if l == 'g':
        train_Y.append(0)
    else:
        train_Y.append(1)
        
for l in test_labels:
    if l == 'g':
        test_Y.append(0)
    else:
        test_Y.append(1)
```

# **Decision Tree**

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_mat, train_Y)
```

```python
#tree.plot_tree(clf.fit(train_mat,train_Y))
```

```python
test_score = accuracy_score(clf.predict(test_mat), test_Y)
train_score = accuracy_score(clf.predict(train_mat), train_Y)
print(train_score)
print(test_score)
```

# SVM

```python
sv = svm.SVC()
sv.fit(train_mat, train_Y)
```

```python
test_score = accuracy_score(sv.predict(test_mat), test_Y)
train_score = accuracy_score(sv.predict(train_mat), train_Y)
print(train_score)
print(test_score)
```

# MLP

**Models**

```python
from keras.models import Model
from keras.layers import Dense
from keras import backend as K
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.layers import BatchNormalization
from keras.optimizers import Adamax
```

***2 Layers (10 Nodes each) ***

```python
def create_2_10():
    model = Sequential()
    model.add(Dense(10, input_shape=(10,)))
    model.add(Dropout(0.2)) 
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    return model    
```

```python
model1 = create_2_10()
model1.summary()
```

**4 Layers (128 Nodes each)**

```python
def create_4_128():
    model = Sequential()
    model.add(Dense(128, input_shape=(10,)))
    model.add(Dropout(0.5)) 
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))
    return model 
```

```python
model1 = create_4_128()
model1.summary()
```

**Optimizers (Choose)**

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model1.fit(np.asarray(train_mat),np.asarray(train_Y),batch_size = 16, epochs = 20)
```

```python
opt = Adamax(lr=0.005, beta_1=0.9, beta_2=0.999)
model1.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
model1.fit(np.asarray(train_mat), np.asarray(train_Y), epochs=500, batch_size=32)
```

```python
test_save = model1.evaluate(np.asarray(test_mat), np.asarray(test_Y))
print(test_save)
```

```python
sgd = optimizers.SGD(lr=0.0006, momentum=0.0001, decay=0.0, nesterov=True)
model1.compile(optimizer=sgd, loss='binary_crossentropy', metrics=['accuracy'])
model1.fit(np.asarray(train_mat),np.asarray(train_Y),batch_size = 16, epochs = 400)
```

```python
test_save = model1.evaluate(np.asarray(test_mat), np.asarray(test_Y))
print(test_save)
```
